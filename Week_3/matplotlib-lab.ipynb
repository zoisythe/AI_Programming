## 练习 M1：绘制简单折线图

**要求：** 根据一周的温度数据绘制一条**红色虚线**折线图，带**圆形标记**，并添加标题和轴标签。

**需要填写的知识点：**
- `plt.plot()` 的参数：
  - `color`：颜色。常用值：`"r"`=红色, `"b"`=蓝色, `"g"`=绿色, `"k"`=黑色
  - `linestyle`：线型。`"-"`=实线, `"--"`=虚线, `"-."` =点划线, `":"`=点线
  - `marker`：标记形状。`"o"`=圆形, `"s"`=方形, `"^"`=三角, `"*"`=星形
- 辅助函数：`plt.title()` 设置标题，`plt.xlabel()` 设置 x 轴标签，`plt.ylabel()` 设置 y 轴标签

```python
# ============================================================
# 练习 M1：绘制简单折线图
# ============================================================
# 目标：使用 plt.plot() 绘制折线图
# 要求：红色虚线、圆形标记、添加标题和轴标签
# ============================================================

import matplotlib.pyplot as plt  # 导入 matplotlib 的 pyplot 模块，约定别名为 plt

# ① 准备数据
days = [1, 2, 3, 4, 5, 6, 7]            # x 轴数据：一周的天数（周一到周日）
temps = [22, 25, 19, 24, 28, 26, 23]    # y 轴数据：每天的温度（摄氏度）

# ② 调用 plt.plot() 绘制折线
# 请在 _____ 处填入正确的参数值：
#   color 应填红色的缩写
#   linestyle 应填虚线的符号
#   marker 应填圆形标记的符号
plt.plot(days, temps, color='r', linestyle='--', marker='o')

# ③ 美化图表：添加标题和轴标签
# 请在 _____ 处填入正确的函数名：
#   设置标题用什么函数？
#   设置 x 轴标签用什么函数？
#   设置 y 轴标签用什么函数？
plt.title("Temperature Change")
plt.xlabel("Day")
plt.ylabel("Temp (°C)")

# ④ 展示图表
plt.show()
```

## 练习 M2：绘制柱状图

**要求：** 根据水果销量数据绘制柱状图，柱体为**天蓝色**、**黑色边框**，添加标题和轴标签，并显示**网格线**。

**需要填写的知识点：**
- `plt.bar(x, height)` 用于绘制柱状图
  - `color`：柱体填充颜色（支持颜色名称如 `"skyblue"`, `"coral"`, `"gold"` 等）
  - `edgecolor`：柱体边框颜色
- `plt.grid(True)` 显示网格线，方便读数

```python
# ============================================================
# 练习 M2：绘制柱状图
# ============================================================
# 目标：使用 plt.bar() 绘制柱状图
# 要求：天蓝色填充、黑色边框、显示网格线
# ============================================================

import matplotlib.pyplot as plt

# ① 准备数据
fruits = ["Apple", "Banana", "Orange", "Grape", "Mango"]  # x 轴：水果名称
sales = [120, 85, 150, 65, 95]                              # y 轴：各水果的销量

# ② 调用画图函数绘制柱状图
# 请在 _____ 处填入正确的内容：
#   第一个 _____ 应填绘制柱状图的函数名
#   color 应填天蓝色的颜色名称
#   edgecolor 应填黑色的颜色名称
plt.bar(fruits, sales, color='skyblue', edgecolor='black')

# ③ 美化图表
plt.title("Fruit Sales")     # 设置标题
plt.xlabel("Fruit")          # 设置 x 轴标签
plt.ylabel("Sales")          # 设置 y 轴标签

# 请在 _____ 处填入显示网格线的函数名
plt.grid(True)

# ④ 展示图表
plt.show()
```

## 练习 M3：绘制散点图

**要求：** 绘制一组学生身高与体重的散点图，点为**绿色三角形**、大小为 **80**、透明度 **0.6**。

**需要填写的知识点：**
- `plt.scatter(x, y)` 用于绘制散点图，常用参数：
  - `c`：点的颜色（如 `"green"`, `"red"` 等）
  - `marker`：点的形状（`"o"`=圆形, `"s"`=方形, `"^"`=三角形, `"*"`=星形）
  - `s`：点的大小（数值，默认约 20，数值越大点越大）
  - `alpha`：透明度（0=完全透明, 1=完全不透明），适合观察重叠数据点

```python
# ============================================================
# 练习 M3：绘制散点图
# ============================================================
# 目标：使用 plt.scatter() 绘制散点图
# 要求：绿色三角形标记、大小 80、透明度 0.6
# ============================================================

import matplotlib.pyplot as plt

# ① 准备数据：10 个学生的身高和体重
height = [155, 160, 165, 170, 175, 180, 168, 172, 158, 163]  # x 轴：身高（cm）
weight = [50, 55, 60, 65, 70, 78, 62, 68, 52, 57]            # y 轴：体重（kg）

# ② 调用画图函数绘制散点图
# 请在 _____ 处填入正确的内容：
#   第一个 _____ 应填绘制散点图的函数名
#   c 应填绿色的颜色名称
#   marker 应填三角形的标记符号
#   s 应填点的大小数值
#   alpha 应填透明度数值
plt.scatter(height, weight, c='green', marker='^', s=80, alpha=0.6)

# ③ 美化图表
plt.title("Height vs Weight")    # 标题
plt.xlabel("Height (cm)")        # x 轴标签
plt.ylabel("Weight (kg)")        # y 轴标签

# ④ 展示图表
plt.show()
```

## 练习 M4：绘制饼图

**要求：** 绘制编程语言受欢迎程度饼图，显示**百分比**（保留 1 位小数），从 **90 度**开始绘制。

**需要填写的知识点：**
- `plt.pie(sizes)` 用于绘制饼图，常用参数：
  - `labels`：各扇区的标签列表
  - `autopct`：百分比显示格式
    - `"%1.1f%%"` 表示保留 1 位小数的百分比
    - 其中 `%1.1f` 是浮点数格式，`%%` 用于显示百分号 `%`
  - `startangle`：饼图起始角度（逆时针方向，90 = 从正上方开始）

```python
# ============================================================
# 练习 M4：绘制饼图
# ============================================================
# 目标：使用 plt.pie() 绘制饼图
# 要求：显示百分比（1 位小数）、从 90 度开始
# ============================================================

import matplotlib.pyplot as plt

# ① 准备数据
languages = ["Python", "Java", "C++", "JavaScript"]  # 编程语言名称（将作为饼图各扇区的标签）
popularity = [35, 25, 20, 20]                          # 各语言的受欢迎程度数值
# 注意：matplotlib 会自动将数值转换为百分比，不需要手动计算

# ② 调用画图函数绘制饼图
# 请在 _____ 处填入正确的内容：
#   第一个 _____ 应填绘制饼图的函数名
#   labels 应填标签列表的变量名
#   autopct 应填百分比格式字符串
#   startangle 应填起始角度的数值
plt.pie(popularity, labels=languages, autopct='%.1f%%', startangle=90)

# ③ 美化图表
plt.title("Programming Language Popularity")

# ④ 展示图表
plt.show()
```

## 练习 M5：子图绘制（进阶）

**要求：** 在同一画布上创建 **1 行 2 列**的子图，左图为**折线图**，右图为**柱状图**。

**需要填写的知识点：**
- `plt.figure(figsize=(w, h))` 创建画布
  - `figsize` 接收一个**元组** `(宽, 高)`，单位是英寸
- `plt.subplot(nrows, ncols, index)` 在画布上创建子图
  - `nrows`：子图的行数
  - `ncols`：子图的列数
  - `index`：当前子图的编号（**从 1 开始**，按从左到右、从上到下的顺序编号）
  - 例如 `plt.subplot(1, 2, 1)` = 1行2列布局中的第 1 个（左边）
- `plt.tight_layout()` 自动调整子图间距，防止标题/标签重叠

```python
# ============================================================
# 练习 M5：子图绘制（进阶）
# ============================================================
# 目标：使用 plt.subplot() 在一张画布上绘制多个子图
# 要求：1 行 2 列，左边折线图，右边柱状图
# ============================================================

import matplotlib.pyplot as plt

# ① 准备数据
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]   # 左图数据（1, 4, 9, 16, 25 是平方数列：1², 2², 3², 4², 5²）
y2 = [5, 3, 8, 2, 7]     # 右图数据

# 创建画布，指定尺寸
# 请在 _____ 处填入画布尺寸的元组（宽 10 英寸，高 4 英寸）
plt.figure(figsize=(10, 4))

# ---- 左图：折线图 ----
# 请在 _____ 处填入创建子图的函数名
# 参数含义：1 行 2 列布局，当前是第 1 个子图（左边）
plt.subplot(1, 2, 1)
plt.plot(x, y1, color="blue", marker="s")  # 蓝色折线，方形标记
plt.title("Line Chart")                     # 左图标题

# ---- 右图：柱状图 ----
# 请在 _____ 处填入创建子图的函数名
# 参数含义：1 行 2 列布局，当前是第 2 个子图（右边）
plt.subplot(1, 2, 2)
plt.bar(x, y2, color="coral")              # 珊瑚色柱状图
plt.title("Bar Chart")                      # 右图标题

# 自动调整子图间距，防止标题和标签互相重叠
plt.tight_layout()

# ④ 展示图表
plt.show()
```
